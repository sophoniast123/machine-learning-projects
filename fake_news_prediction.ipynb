Fake News Prediction 

0 --> fake news
1 --> real news


```python
import numpy as np
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
nltk.download('stopwords')
```

Data preprocessing


```python
news_dataset = pd.read_csv('train (1).csv')
```

```python
news_dataset.shape
```

```python
news_dataset.head()
```

```python
news_dataset.isnull().sum()
```

```python
news_dataset = news_dataset.fillna('')
```

```python
news_dataset.isnull().sum()
```

```python
news_dataset['content'] = news_dataset['author']+ " " +news_dataset['title']
```

```python
x = news_dataset.drop(columns='label', axis= 1)
y = news_dataset['label']
```

```python
port_stem = PorterStemmer()
```

```python
def steamming(content):
    steammed_content = re.sub(r'[^a-zA-Z]', " ", content)
    steammed_content = steammed_content.lower()
    steammed_content = steammed_content.split()
    steammed_content = [port_stem.stem(word) for word in steammed_content if word not in stopwords.words('english')]
    steammed_content = " ".join(steammed_content)
    return steammed_content
```

```python
news_dataset['content'] = news_dataset['content'].apply(steamming)
```

```python
print(news_dataset['content'])
```

```python
x = news_dataset['content'].values
y = news_dataset['label'].values
```

```python
print(y)
```

vectorize bc the maching can't understands words 

words --> numbers by tfidfvectorizer

```python
vectorizer = TfidfVectorizer()
```

```python
vectorizer.fit(x)

x = vectorizer.transform(x)
```

splitting the dataset to train and test

```python
x_train, x_test, y_train, y_test = train_test_split(x ,y ,test_size=0.2, stratify = y, random_state=2)
```

Training the model: using logostic regression model 

```python
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

Evaluation

Accuracy Score


```python
x_train_predication = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_predication, y_train)
```

```python
print('accuracy score of the training data : ', training_data_accuracy )
```

```python
x_test_predication = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_predication, y_test)
```

```python
print('accuracy score of the test data : ', test_data_accuracy )
```

Making a predictive System

```python
x_new = x_test[1]
prediction = model.predict(x_new)
print(prediction)

if (prediction[0]==0):
    print('the news is real')
else:
    print('the news is fake')
```
